# q1 单字统计

```python
from collections import defaultdict
```

```python
path = 'words.txt'
with open(path, 'r') as f:
    text = f.read().strip().split(' ')
text[:5]
```

```python
word_cnt = dict((w, text.count(w)) for w in text)
```

```python
wd = []
for w in text :
    if w not in wd:
        wd.append(w)
```

```python
with open('Q1.txt' , 'w') as f:
    idx  = 0
    for w in wd[:-1]:
        f.write(w+' '+str(idx)+' '+str(word_cnt[w])+'\n')
        idx+=1
    f.write(w+' '+str(idx)+' '+str(word_cnt[w]))
```

# Q2图片淡化

```python
import cv2
```

```python
img = cv2.imread('Q2.jpg')
img
```

```python
import numpy as np
img = np.array(img)
```

```python
img.shape
```

```python
img[1,1,:]
```

```python
for i in range(img.shape[0]):
    for j in range(img.shape[1]):
        b,g,r = img[i,j,:]
        img[i,j] = (r//2,g//2,b//2)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
plt.imshow(img)
```

```python
plt.imsave('westbrook2.png',img)
```
