```python
import sys
import numpy as np
from keras.models import Sequential
from keras.layers.core import Dense, Activation
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
seed = 777
np.random.seed(seed)
```

```python
x = np.linspace(-1,1,1000)
y = np.sinc(x)

plt.plot(x,y)
```

```python
model_0 = Sequential([
    Dense(4,input_shape=(1,),activation='selu',kernel_initializer='normal'),
    Dense(8,activation='selu',kernel_initializer='normal'),
    Dense(8,activation='selu',kernel_initializer='normal'),
    Dense(16,activation='selu',kernel_initializer='normal'),
    Dense(8,activation='selu',kernel_initializer='normal'),
    Dense(8,activation='selu',kernel_initializer='normal'),
    Dense(4,activation='selu',kernel_initializer='normal'),
    Dense(1,kernel_initializer='normal'),
])
model_0.summary()  #8å±‚
```

```python
model_1 = Sequential([
    Dense(4,input_shape=(1,),activation='selu',kernel_initializer='normal'),
    Dense(14,activation='selu',kernel_initializer='normal'),
    Dense(27,activation='selu',kernel_initializer='normal'),
    Dense(1,kernel_initializer='normal'),
])
model_1.summary()
```

```python
model_2 = Sequential([
    Dense(2,input_shape=(1,),activation='selu',kernel_initializer='normal'),
    Dense(127,activation='selu',kernel_initializer='normal'),
    Dense(1,kernel_initializer='normal'),
])
model_1.summary()
```

```python
batch_size = 128
epochs = 1000
i = 0
for model in [model_0, model_1, model_1]:
    model.compile(loss='mse', optimizer='adam')
    fitHistory = model.fit(x, y, epochs=epochs, batch_size=batch_size, shuffle = True)
    model.save('Sinc_{}.h5'.format(i))
    np.save('train_loss_history_Sinc_{}.npy'.format(i), fitHistory.history['loss'])
    i+=1
```

```python
sinc_0 = np.load('train_loss_history_Sinc_0.npy')
sinc_1 = np.load('train_loss_history_Sinc_1.npy')
sinc_2 = np.load('train_loss_history_Sinc_2.npy')
    

plt.plot(sinc_0)
plt.plot(sinc_1)
plt.plot(sinc_2)
plt.title('Loss v.s. Epoch')
plt.ylabel('Loss')
plt.yscale('log')
plt.xlabel('# of epoch')
plt.legend(['model0_loss', 'model1_loss', 'model2_loss'],loc='lower left')
plt.savefig('SINC_LOSS.png')
    
```

```python
from keras.models import load_model
model_0 = load_model('Sinc_0.h5')
model_1 = load_model('Sinc_1.h5')
model_2 = load_model('Sinc_2.h5')

x_test = np.linspace(-1,1,1000)
y_test = np.sinc(x_test)

predict_0 = np.squeeze(model_0.predict(x_test))
predict_1 = np.squeeze(model_1.predict(x_test))
predict_2 = np.squeeze(model_2.predict(x_test))

plt.plot(x_test, y_test, color='black')
plt.plot(x_test, predict_0, color='red')
plt.plot(x_test, predict_1, color='green')
plt.plot(x_test, predict_2, color='blue')
plt.title('Prediction')
plt.ylabel('Y')
plt.xlabel('X')
plt.legend(['sinc(x)', 'model0', 'model1', 'model2'],
loc='upper right')
plt.savefig('SINC_Predict.png')
```
